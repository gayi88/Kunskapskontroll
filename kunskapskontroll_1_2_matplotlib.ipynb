### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
#1. Explicit interface : Creates Figures and Axes, and call methods on them  (object oriented interface)
#2.Implicit interface: It  uses pyplot functions for plotting.This interface keep track of the Axes created,and adds Artists to the object it thinks the user wants.
#3. In both interfaces, Explicit interface is recommended because of its flexibility and underlies the other interfaces
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python

#Figure:overall place where we are creating any kind of charts or plots. Figure have multiple axes (titles,figure legends,colorbars and etc)
#Axes: its an Artist attached to a Figure that contains a region for plotting data,and usually includes 2 axis objects that provide ticks and tick labels to provide scales for the data in the axes.
#axis:Axis is reference lines for the plot, for both x-axis and y-axis.These axes define the coordinate system used for plotting the data.
#Artist: everything visible on the Figure is called a Artist(all Figures, Axes and Axis).It is the base classs for everything visual.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x= np.linspace(-4,4)
y =[i**2 for i in x ]
plt.plot(x,y)
plt.title("Function plot")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.show()


```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,6))
ax1.scatter(x,y)


# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())


ax2.bar(names,values)
ax2.set_title('Bar plot')
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Quantity')

plt.tight_layout()
plt.show()

```
